**Desafio 01: Ordenar o gráfico countplot**

**Desafio 02: Melhorar a visualização alterando tamanho da fonte**

```python
plt.figure(figsize=(10.5, 8))
plt.rc('axes', titlesize=14)    
plt.rc('axes', labelsize=14)   
plt.rc('xtick', labelsize=14)    
plt.rc('ytick', labelsize=14)    
ax = sns.countplot(x = 'composto', data=df.query('composto in @compostos'), order= compostos)
```

**Desafio 03: Plotar os histogramas com seaborn**

```python
plt.figure(figsize=(10,8))

x = np.random.normal(size = 2000)
plt.hist(df["g-0"], bins=30, color='#00bcd4' , edgecolor="#6A9662")
plt.show()

plt.figure(figsize=(10,8))
plt.hist(df["g-1"], bins=20, color='blue' , edgecolor="#6A9662")
plt.show()


plt.figure(figsize=(10,8))
plt.hist(df["g-2"], bins=20, color='green' , edgecolor="#6A9662")
plt.show()
plt.savefig("photo-1.1.PNG", format='png')
```

**Desafio 04: Fazer outras análises com o boxplot e até com o histograma.**

```python
plt.subplots(figsize = (14,24))
plt.subplot(3,1,1)

sns.boxplot(x= df['g-0'] , y= df['dose'] ,  width=.7 )

plt.subplot(3,1,2)
sns.boxplot(x= df['g-1'] , y= df['dose'] ,  width=.5)

plt.subplot(3,1,3)
sns.boxplot(x= df['g-2'] , y= df['dose'] ,  width=.25)

plt.show()
plt.savefig("photo-1.2.PNG", format='png')
```
